```python
import PySimpleGUI as sg

"""
    Demo - Class wrapper
    
    Using a class to encapsulate PySimpleGUI Window creation & event loop

    Copyright 2022 PySimpleGUI
"""


class SampleGUI():

    def __init__(self):
        self.layout = [ [sg.Text('My layout')],
                        [sg.Input(key='-IN-')],
                        [sg.Button('Go'), sg.Button('Exit')] ]

        self.window = sg.Window('My new window', self.layout)

    def run(self):
        while True:             # Event Loop
            self.event, self.values = self.window.read()
            if self.event in (sg.WIN_CLOSED, 'Exit'):
                break

            if self.event == 'Go':
                self.button_go()

        self.window.close()

    def button_go(self):
        sg.popup('Go button clicked', 'Input value:', self.values['-IN-'])

# Create the class
my_gui = SampleGUI()
# run the event loop
my_gui.run()
```

```python
import PySimpleGUI as sg

class InputTest:
    """AI is creating summary for 
    test the input
    """

    def __init__(self) -> None:
        """AI is creating summary for __init__
        init the window
        """
        self. data_of_window = [[sg.Text('enter email address to check')],
                                [sg.Input(key='-EL-')],
                                [sg.Text('enter ID to check')],
                                [sg.Input(key='-ID-')],
                                [sg.Button('Check'), sg.Button('Exit')]]
        self.window = sg.Window('Checker email address and ID', self.data_of_window)

    def run(self) -> None:
        """loop of window, take input then check
        """
        while True:
            self.event, self.values = self.window.read()
            if self.event in (sg.WIN_CLOSED, 'Exit'):
                break
            if self.event == 'Check':
                self.Responds_to_user()
        self.window.close()

    def Responds_to_user(self) -> None:
        """Responds to the user if input correct or not
        """
        sg.popup('email and ID checked',
                 'input email:', self.values['-EL-'],
                 'Input ID:', self.values['-ID-'],
                 'Valid email address' if EL_is_correct(self.values['-EL-']) else 'Invalid email address',
                 'Valid ID' if ID_is_correct(self.values['-ID-']) else 'Invalid ID')

def ID_is_correct(ID: str) -> bool:
    """AI is creating summary for ID_is_correct
    check ID is correct or not

    Args:
        ID (str): [ID to check]

    Returns:
        bool: [ID is correct or not]
    """
    if len(ID) < 9:
        return False
    numbers_id = [
        sum(
            int(digit)
            for digit in str(
                (int(ID[index]) * 2 - (int(ID[index]) * ((index + 1) % 2)))
            )
        )
        for index in range(len(ID) - 1)
    ] # array of calculated numbers according to the calculation method
    return int(ID[-1]) == abs(10 - sum(numbers_id) % 10) % 10

def EL_is_correct(EL: str) -> bool:
    """AI is creating summary for EL_is_correct
    check EL is correct or not

    Args:
        ID (str): [EL to check]

    Returns:
        bool: [EL is correct or not]
    """
    if len(EL) < 4:
        return False
    if EL[0] == '@' or EL[-1] == '@' or EL[-1] == '.' or EL.count('@') != 1:
        return False
    return EL.split('@')[1][0] != '.' and '.' in EL.split('@')[1]

my_gui = InputTest()
my_gui.run()
```

```python
import webbrowser

class SearchLocal:
    """AI is creating summary for SearchLocal
    """
    def __init__(self) -> None:
        """AI is creating summary for __init__
        init window
        """
        self. data_of_window = [[sg.Text('enter local to search')],
                                [sg.Input(key='-LOCAL-')],
                                [sg.Button('Search'), sg.Button('Exit')]]
        self.window = sg.Window('Search local in google maps', self.data_of_window)

    def run(self) -> None:
        """loop of window, take input then check
        """
        while True:
            self.event, self.value = self.window.read()
            if self.event in (sg.WIN_CLOSED, 'Exit'):
                break
            if self.event == 'Search':
                webbrowser.open(build_url_for_maps(self.value['-LOCAL-']))
        self.window.close()

def build_url_for_maps(local: str) -> str:
    """AI is creating summary for build_url_for_maps
    take words to search then return url to search in google maps

    Args:
        local (str): [words to search]

    Returns:
        str: [url to search in google maps]
    """
    words_to_search = local.split()
    url_search = "https://www.google.co.il/maps/search/"
    for index in range(len(words_to_search) -1):
        url_search += f'{words_to_search[index]}+'
    url_search += words_to_search[-1]
    url_search += "%E2%80%AD/@31.7861008,35.268665,12z/data=!3m1!4b1"
    return url_search

search = SearchLocal()
search.run()

```
