```python
"""Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the negative numbers red and positive numbers black.

Write function color_negative_red()
"""

import pandas as pd
import numpy as np
np.random.seed(24)

df = pd.DataFrame({"A": np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list("BCDE"))], axis=1)
print("Original array:")
print(df)

def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return f'color: {color}'

print("\nNegative numbers red and positive numbers black:")
df.style.applymap(color_negative_red)
```

```python
"""Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the maximum value in last two columns."""

import pandas as pd
import numpy as np
np.random.seed(24)
df = pd.DataFrame({"A": np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list("BCDE"))], axis=1)
df.iloc[0, 2] = np.nan
df.iloc[3, 3] = np.nan
df.iloc[4, 1] = np.nan
df.iloc[9, 4] = np.nan
print("Original array:")
print(df)


def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: green' if v else '' for v in is_max]

print("\nHighlight the maximum value in last two columns:")
df.style.apply(highlight_max,subset=pd.IndexSlice[:, ['D', 'E']])
```
