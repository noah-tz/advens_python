```python
"""(1) Write a Pandas program to create and display a DataFrame from a specified dictionary data which has the index labels.

Sample Data:
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

OUTPUT:
        name  score  attempts qualify
a  Anastasia   12.5         1     yes
b       Dima    9.0         3      no
c  Katherine   16.5         2     yes
d      James    NaN         3      no
e      Emily    9.0         2      no
f    Michael   20.0         3     yes
g    Matthew   14.5         1     yes
h      Laura    NaN         1      no
i      Kevin    8.0         2      no
j      Jonas   19.0         1     yes
"""
import pandas as pd
import numpy as np

exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data, index=labels)
df
```

```python
"""(2)
Write a Pandas program to select the specified columns and rows from a given DataFrame.
Select 'name' and 'score' columns in rows 1, 3, 5, 6 from the following data frame.

OUTPUT:
Select specific columns and rows:
   score qualify
b    9.0      no
d    NaN      no
f   20.0     yes
g   14.5     yes
"""
print(df.iloc[[1, 3, 5, 6], [1, 3]])
```

```python
"""(3)
Write a Pandas program to select the rows where the score is missing, i.e. is NaN.


OUTPUT:
Rows where score is missing:
    name  score  attempts qualify
d  James    NaN         3      no
h  Laura    NaN         1      no
"""
df[df['score'].isnull()]
```

```python
"""(4) Write a Pandas program to sort the data frame first by 'name' in descending order, then by 'score' in ascending order.

OUTPUT:
Orginal rows:
        name  score  attempts qualify
a  Anastasia   12.5         1     yes
b       Dima    9.0         3      no
c  Katherine   16.5         2     yes
d      James    NaN         3      no
e      Emily    9.0         2      no
f    Michael   20.0         3     yes
g    Matthew   14.5         1     yes
h      Laura    NaN         1      no
i      Kevin    8.0         2      no
j      Jonas   19.0         1     yes

Sort the data frame first by ‘name’ in descending order, then by ‘score’ in ascending order:
        name  score  attempts qualify
a  Anastasia   12.5         1     yes
b       Dima    9.0         3      no
c  Katherine   16.5         2     yes
d      James    NaN         3      no
e      Emily    9.0         2      no
f    Michael   20.0         3     yes
g    Matthew   14.5         1     yes
h      Laura    NaN         1      no
i      Kevin    8.0         2      no
j      Jonas   19.0         1     yes
"""
df2 = df.copy()
df2['name'][8] = 'Anastasia'
df2.sort_values(by=['name', 'score'], ascending=[False, True])
```

```python
"""(5) Write a Pandas program to change the order of a DataFrame columns.

OUTPUT:
Original DataFrame
   col1  col2  col3
0     1     4     7
1     4     5     8
2     3     6     9
3     4     7     0
4     5     8     1

After altering col1 and col3
   col3  col2  col1
0     7     4     1
1     8     5     4
2     9     6     3
3     0     7     4
4     1     8     5
"""
df_dict = {'col1': [1, 4, 3, 4, 5],
           'col2': [4, 5, 6, 7, 8],
           'col3': [7, 8, 9, 0, 1]}

labels_rows = [0, 1, 2, 3, 4]

df_numbers = pd.DataFrame(df_dict, labels_rows)
df_numbers[['col3', 'col2', 'col1']]
```

```python
"""(6) Write a Pandas program to drop a list of rows from a specified DataFrame.

OUTPUT:
Original DataFrame
   col1  col2  col3
0     1     4     7
1     4     5     8
2     3     6     9
3     4     7     0
4     5     8     1

New DataFrame after removing 2nd & 4th rows:
   col1  col2  col3
0     1     4     7
1     4     5     8
3     4     7     0
"""
print(df_numbers.iloc[[0, 1, 3], [0, 1, 2]])
```

```python
"""(7) Write a Pandas program to remove infinite values from a given DataFrame.

OUTPUT:
Original DataFrame:
        0
0  1000.0
1  2000.0
2  3000.0
3 -4000.0
4     inf
5    -inf

Removing infinite values:
        0
0  1000.0
1  2000.0
2  3000.0
3 -4000.0
4     NaN
5     NaN

""" 
numbers = [1000, 2000, 3000, -4000, np.inf, -np.inf]
df_numbers = pd.DataFrame(numbers)
print(f"original array is\n{df_numbers}\n")
print(f"after replace is\n{df_numbers.replace([np.inf, -np.inf], np.nan)}\n")
```

```python
"""(8) Write a Pandas program to split a given dataframe into groups and create a new column with count from GroupBy.

OUTPUT:
Original Orders DataFrame:
  book_name book_type  book_id
0     Book1      Math        1
1     Book2   Physics        2
2     Book3  Computer        3
3     Book4   Science        4
4     Book1      Math        1
5     Book2   Physics        2
6     Book3  Computer        3
7     Book5   English        5

New column with count from groupby:
  book_name book_type  count
0     Book1      Math      2
1     Book2   Physics      2
2     Book3  Computer      2
3     Book4   Science      1
4     Book5   English      1
"""

book_name = ['Book1','Book2','Book3','Book4','Book1','Book2','Book3','Book5']
book_type = ['Math','Physics','Computer','Science','Math','Physics','Computer','English']
book_id = [1,2,3,4,1,2,3,5]
dict_books = {'book_name': book_name, 'book_type': book_type, 'book_id': book_id}
df_books = pd.DataFrame(dict_books)
print(f'original data is\n{df_books}\n')
print('result is')
print(df_books.groupby(["book_name", "book_type"])["book_type"].count().reset_index(name="count"))
```
