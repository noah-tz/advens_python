```python
import random
from datetime import datetime, timedelta

class ParkingTicket:
    def __init__(self, car_plate_number, entry_datetime, exit_datetime, company, parking_slot_index, car_color):
        self.car_plate_number = car_plate_number
        self.entry_datetime = entry_datetime
        self.exit_datetime = exit_datetime
        self.company = company
        self.parking_slot_index = parking_slot_index
        self.car_color = car_color


def generate_random_ticket():
    plate_number_len = random.randint(8,10)
    car_plate_number = ''.join([random.choice('0123456789') for i in range(plate_number_len)])
    entry_datetime = datetime.now() - timedelta(days=random.randint(0, 7), 
                                                hours=random.randint(0, 23), 
                                                minutes=random.randint(0, 59)
                                               )
    exit_datetime = entry_datetime + timedelta(hours=random.randint(1, 50),
                                               minutes=random.randint(0, 59)
                                              )
    company = random.choice(['Amazon','Alphabet','Facebook','Microsoft','Apple','Alibaba' ,'Tesla'])
    parking_slot_index = random.randint(1, 100)
    car_color = random.choice(['Red', 'Blue', 'Green', 'Black', 'White'])
    
    return ParkingTicket(car_plate_number, entry_datetime, exit_datetime, company, parking_slot_index, car_color)

tickets = [generate_random_ticket() for _ in range(50)]
```

```python
vars(tickets[0])
```

```python
df = pd.DataFrame([vars(t) for t in tickets])
```

```python
df.head()
```

```python
df['duration'] = df['exit_datetime'] - df['entry_datetime']
```

```python
df['duration_in_hours'] = df.duration.apply(lambda x:x.total_seconds()/60/60)
```
